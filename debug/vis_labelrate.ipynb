```python
def draw_ratio_conf(ratio_list, conf_r1, conf_r2, conf_r3, title, x_label, save_path='line_graph.png'):
    """
    绘制折线图用于比较ratio-Confidence，并保存图像。

    Args:
        ratio_list (list): 横轴 ratio 的变化
        conf_dataset (list): Confidence 值
        title (str): title
        save_path (str): 保存图像的路径和文件名
    """
    plt.figure(figsize=(10, 6))
    
    plt.plot(ratio_list, conf_r1, marker='o', label='label-20', color='darkslateblue', linestyle='-')
    plt.plot(ratio_list, conf_r2, marker='o', label='label-40', color='darkcyan', linestyle='-')
    plt.plot(ratio_list, conf_r3, marker='o', label='label-60', color='darkmagenta', linestyle='-')
    # plt.plot(ratio_list, ece_cal_pubmed, marker='o', markerfacecolor='white', label='Cal-TS-Pubmed', color='mediumpurple', linestyle='--')
    
    plt.title(title)
    plt.xlabel(x_label, fontsize=16)
    plt.ylabel('Confidence', fontsize=16)
    
    plt.legend(loc='upper left')
    plt.grid(linestyle='-.')

    plt.savefig(save_path)
    print(f"Image save to {save_path}")
    plt.show()
```

```python
def draw_ratio_ece(ratio_list, ece_uncal, ece_cal, title, x_label, save_path='line_graph.png'):
    """
    绘制折线图用于比较ratio-ECE，并保存图像。

    Args:
        ratio_list (list): 横轴 ratio 的变化
        ece_uncal (list): 未校准 ECE 值
        ece_cal (list): 校准 ECE 值
        save_path (str): 保存图像的路径和文件名
    """
    plt.figure(figsize=(10, 6))
    
    plt.plot(ratio_list, ece_uncal, marker='o', label='Uncal-Cora', color='steelblue', linestyle='-')
    plt.plot(ratio_list, ece_cal, marker='o', markerfacecolor='white', label='Cal-TS-Cora', color='steelblue', linestyle='--')
    
#     plt.plot(ratio_list, ece_uncal_citeseer, marker='o', label='Uncal-Citeseer', color='goldenrod', linestyle='-')
#     plt.plot(ratio_list, ece_cal_citeseer, marker='o', markerfacecolor='white', label='Cal-TS-Citeseer', color='goldenrod', linestyle='--')
    
#     plt.plot(ratio_list, ece_uncal_pubmed, marker='o', label='Uncal-Pubmed', color='mediumpurple', linestyle='-')
#     plt.plot(ratio_list, ece_cal_pubmed, marker='o', markerfacecolor='white', label='Cal-TS-Pubmed', color='mediumpurple', linestyle='--')
    
    plt.title(title)
    plt.xlabel(x_label, fontsize=16)
    plt.ylabel('ECE', fontsize=16)
    
    plt.legend(loc='upper right')
    plt.grid(linestyle='-.')

    plt.savefig(save_path)
    print(f"Image save to {save_path}")
    plt.show()
```
